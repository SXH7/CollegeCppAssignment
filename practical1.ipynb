```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset("titanic")
print(df.head(5))
```

```python
print("\n Dataset Info :")
df.info()
```

```python
print("\n Dataset Description (numeric columns):")
print(df.describe())
```

```python
nominal_vars = ["sex","class","embarked","who","deck","alone"]
print("\nNominal data :")
for col in nominal_vars:
 print(f"\ncolumn: {col}")
 print(df[col].value_counts())
```

```python
ordinal_var = "class"
print(f"\nOrdinal data : {ordinal_var}")
print(df[ordinal_var].value_counts().sort_index())
sns.countplot(x=ordinal_var, data = df , order = [ "first","second","third"])
plt.title("passenger count by class (ordinal)")
plt.show()
```

```python
interval_var = "age"
print(f"\nInterval data : {interval_var}")
print(df[interval_var].describe())
sns.histplot(df[interval_var].dropna(),kde = True)
plt.title("Distribution of age (interval data)")
plt.show()
```

```python
ratio_vars = ["fare","age","sibsp","parch"]
print("\nRatio data summary :")
print(df[ratio_vars].describe())
```

```python
sns.boxplot(x="class", y="fare", data=df, order=["First", "Second", "Third"])
plt.title("Fare Distribution by Passenger Class")
plt.show()
```

```python
sns.histplot(df["fare"], bins=30, kde=True)
plt.title("Distribution of Fare (Ratio Variable)")
plt.show()
```
