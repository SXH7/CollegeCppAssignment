```python
import numpy as np
from scipy import linalg as LA
```

```python
A = np.array([[4.0,3.0],[2.0,2.0],[-1.0,-3.0],[-5.0,-2.0]])
U , S , Vt = LA.svd(A)
print(U)
print(S)
print(Vt)
X = np.array([0.243,0.97])
X = X/LA.norm(X)
P = np.dot(Vt, X)
print(P)
S = LA.diagsvd(S,4,2)
Q = S@P
print("Q values",Q)
y= U@Q
print("values of y",y)
print("product of A*X",A@X)
```
