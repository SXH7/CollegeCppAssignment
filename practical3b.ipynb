```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
data = {
"study_hours": [ 2,3,4,5,6,7,8,9,10],
"practice_tests": [1,1,2,2,3,3,4,4,5],
"sleep_hours":[8,7,6,7,6,6,5,5,4],
"exam_score":[35,40,45,55,60,65,70,78,85]
}
df = pd.DataFrame(data)
# independent variables
x = df[["study_hours","practice_tests","sleep_hours"]]
# dependent variable
y = df["exam_score"]
```

```python
model=LinearRegression()
model.fit(x,y)
# Get coefficients
intercept=model.intercept_
coefficients=model.coef_
print("Intercept (b0):", intercept)
print(f"Equation: y= {intercept:.2f}+{coefficients[0]:.2f}*study_hours"f"+ {coefficients[1]:.2f}*Practice_tests +{coefficients[2]:.2f}*sleep_hours")
```

```python
# Predictions
y_pred = model.predict(x)
# Model evalution
print(f"mean square error:", mean_squared_error(y, y_pred))
print(f"R-squared:",r2_score( y, y_pred))
```

```python
# visualizations
plt.scatter(y , y_pred , color = "blue")
plt.xlabel("Actual exam score")
plt.ylabel("predicted exam score")
plt.title("Multiple Linear Regression : Actual v/s predicted")
plt.plot([y.min(),y.max()],[y.min(), y.max()], color = "red", linestyle="--")
plt.show()
```
