```python
import numpy as np
from scipy import linalg as LA
import matplotlib.pyplot as plt
D= np.array([[0,4,3,7,8],[4,0,1,6,7],[3,1,0,5,7],[7,6,5,0,1],[8,7,7,1,0]])
D2 = np.square(D)
C = np.eye(5)-0.2*np.ones(5)
M = -0.5*C@D2@C
l,v = LA.eig(M)
s = np.real(np.power(l,0.5))
V2 = v[:,[0,1]]
s2 = np.diag(s[0:2])
Q = V2 @ s2
plt.plot(Q[:,0],Q[:,1],"ro")
plt.show()
```
